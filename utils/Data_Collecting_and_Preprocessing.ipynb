```python
import os
import cv2
```

```python
inp_folder = r"E:\data"
out_train_folder = r"D:\Face Segmentation\data\images\train"
out_test_folder = r"D:\Face Segmentation\data\images\test"
```

```python
os.makedirs(out_train_folder, exist_ok = True)
for i in range(1,144):
    os.makedirs(out_test_folder + str(i), exist_ok = True)
```

```python
i = 1
for img_file in os.listdir(inp_folder):
    if img_file.endswith(('.jpg', '.png', '.jpeg')):
        img_path = os.path.join(inp_folder, img_file)
        image = cv2.imread(img_path)
        img_name = 'face' + str(i) + '.jpg'

        if (i <= 50):
            output_img_path = os.path.join(out_train_folder, img_name)
            cv2.imwrite(output_img_path, image)
            print(f"Processed and saved: {output_img_path}")

        else:
            output_img_path = os.path.join(out_test_folder + str((i-1)//50), img_name)
            cv2.imwrite(output_img_path, image)
            print(f"Processed and saved: {output_img_path}")
            
        i += 1
```
